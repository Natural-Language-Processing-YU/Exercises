# 1.0 Law of Cosines

$
c^2 = a^2 + b^2 - 2ab \cos(C)
$

In this formula:
- $c$ represents the length of the side opposite to angle $C$.
- $a$ and $b$ represent the lengths of the other two sides.
- $\cos(C)$ denotes the cosine of angle $C$.



## 1.1 Example of Law of Cosines

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
a = 5
b = 7
angle_C = np.deg2rad(45)  # Convert angle to radians

# Calculate the length of the third side using the Law of Cosines
c_squared = a**2 + b**2 - 2*a*b*np.cos(angle_C)
c = np.sqrt(c_squared)

# Plotting the triangle
plt.figure()
plt.plot([0, a], [0, 0], 'b-', label='Side a')
plt.plot([a, a - b * np.cos(angle_C)], [0, b * np.sin(angle_C)], 'r-', label='Side b')
plt.plot([0, a - b * np.cos(angle_C)], [0, b * np.sin(angle_C)], 'g-', label='Side c')
plt.plot([0], [0], 'ko')  # Origin
plt.plot([a - b * np.cos(angle_C)], [b * np.sin(angle_C)], 'ko')  # Vertex
plt.legend()
plt.xlim(-1, 8)
plt.ylim(-1, 8)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Triangle visualization')
plt.grid(True)
plt.show()

# Print the length of side c
print("Length of side c:", c)
```

## 2.0 Dot Product of Two Vectors

The dot product (also known as the inner product or scalar product) of two vectors is an operation that combines their magnitudes and the angle between them. Mathematically, the dot product of two vectors $ \mathbf{a} $ and $ \mathbf{b} $ is defined as:

$$
\mathbf{a} \cdot \mathbf{b} = \|\mathbf{a}\| \|\mathbf{b}\| \cos(\theta)
$$

Where:
- $ \mathbf{a} $ and $ \mathbf{b} $ are vectors.
- $ \|\mathbf{a}\| $ and $ \|\mathbf{b}\| $ represent the magnitudes (or lengths) of vectors $ \mathbf{a} $ and $ \mathbf{b} $, respectively.
- $ \theta $ is the angle between the two vectors.

The dot product can be computed by multiplying corresponding components of the vectors and summing them. If $ \mathbf{a} = [a_1, a_2, a_3] $ and $ \mathbf{b} = [b_1, b_2, b_3] $, then their dot product is given by:

$$
\mathbf{a} \cdot \mathbf{b} = a_1 b_1 + a_2 b_2 + a_3 b_3
$$

The dot product has several important properties, such as commutativity ( $ \mathbf{a} \cdot \mathbf{b} = \mathbf{b} \cdot \mathbf{a} $ ) and distributivity ( $ \mathbf{a} \cdot (\mathbf{b} + \mathbf{c}) = \mathbf{a} \cdot \mathbf{b} + \mathbf{a} \cdot \mathbf{c} $ ).

### 2.1 Vector Length (Magnitude)

The length (or magnitude) of a vector represents its size or magnitude in the vector space. It is calculated using the Euclidean norm (also known as the L2 norm or 2-norm). For a vector $ \mathbf{v} = [v_1, v_2, \ldots, v_n] $, the length (magnitude) is given by:

$$
\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + \ldots + v_n^2}
$$

The length of a vector can be interpreted as the distance from the origin to the point represented by the vector in the vector space.

In some contexts, the squared length of a vector, $ \|\mathbf{v}\|^2 $, is also used and can be computed simply by taking the dot product of the vector with itself ( $ \|\mathbf{v}\|^2 = \mathbf{v} \cdot \mathbf{v} $ ).

The length of a vector is always a non-negative value. If the length is zero ( $ \|\mathbf{v}\| = 0 $ ), the vector is known as the zero vector.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
a = np.array([3, 4])  # Vector a = [3, 4]
b = np.array([1, -2])  # Vector b = [1, -2]

# Calculate the dot product of vectors a and b
dot_product = np.dot(a, b)

# Calculate the lengths of vectors a and b
length_a = np.linalg.norm(a)
print(a)
length_b = np.linalg.norm(b)
print(b)

# Print the dot product and vector lengths
print("Dot Product of a and b:", dot_product)
print("Length of vector a:", length_a)
print("Length of vector b:", length_b)

# Plotting the vectors
plt.figure()
plt.plot([0, a[0]], [0, a[1]], 'r-', label='Vector a')
plt.plot([0, b[0]], [0, b[1]], 'b-', label='Vector b')
plt.legend()
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Visualization of Vectors')
plt.grid(True)
plt.show()
```
